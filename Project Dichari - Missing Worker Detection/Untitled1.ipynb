```python
import os
```

```python
local_path = r"C:\Users\Asus\OneDrive\Desktop\Coded dataset"
count = 0
for file in os.listdir(local_path):
  file_path = os.path.join(local_path,file)
  if file.endswith(".txt") and not(file=="classes.txt"):
    f = open(file_path,'r')
    temp = ""
    for x in f:
      x = x.split(" ")
      if x=='\n' or x=='':
            print(file_path)
      x[0] = 0
      for i in range(1,len(x)):
        x[i] = float(x[i])
      for i in x:
        temp+=str(i)
        temp+=" "
      temp = temp[:-1]
      temp+="\n"
    f = open(file_path,'w')
    f.truncate(0)
    temp = temp.rstrip()
    print(temp)
    f.write(temp)
    f.close()
```
